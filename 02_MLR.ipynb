# [重回帰分析] 面積と築年数から不動産価格を予測する

## (1) 必要な部品を用意する

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

## (2) 学習データを用意する

```python
price = [10000000, 18000000, 27000000, 36000000, 22000000] # 価格
area  = [      60,       84,      200,      110,       80] # 面積
age   = [      28,       22,       14,       10,        8] # 築年数
df = pd.DataFrame(list(zip(price,area,age)), columns = ['price','area', 'age'])
df
```

## (3) グラフで視覚化する

```python
df.plot.scatter('price', 'area')
df.plot.scatter('price', 'age')
```

## (4) 学習する変数を分ける

```python
X = df[['area', 'age']] # 面積と築年数
y = df['price']         # 価格
```

## (5) 機械に学習させる

```python
reg = LinearRegression()
reg.fit(X, y)
```

## (6) 機械に予測させる(面積が80㎡で築年数が10年の不動産価格は?)

```python
tmp0 = pd.DataFrame([[80, 10]],columns = ["area","age"])
tmp1 = reg.predict(tmp0)
tmp2 = int(tmp1[0])
print("{:,} 円".format(tmp2))
```
