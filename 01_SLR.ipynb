# [単回帰分析] 身長から体重を予測する

## (1) 必要な部品を用意する

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

## (2) 学習データを用意する

```python
height = [170.6,171.0,171.3,170.5,171.6,170.6,169.7,170.3,170.4,170.6]
weight = [ 63.4, 65.5, 64.1, 63.5, 65.4, 64.4, 63.8, 63.6, 62.5, 62.7]
df = pd.DataFrame(list(zip(height,weight)), columns = ['height','weight'])
df
```

## (3) グラフで視覚化する

```python
df.plot.scatter('height', 'weight')
```

## (4) 学習する変数を分ける

```python
X = df[['height']]
y = df['weight']
```

## (5) 機械に学習させる

```python
reg = LinearRegression()
reg.fit(X, y)
```

## (6) 機械に予測させる(身長が175.0の人の体重は?)

```python
tmp0 = pd.DataFrame([[175.0]],columns = ["height"])
tmp1 = reg.predict(tmp0)
tmp2 = int(tmp1[0])
print("{:,} kg".format(tmp2))
```
